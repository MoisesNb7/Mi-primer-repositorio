Tipo de datos genericos de NumPy como number

```python
import numpy as np

def subtypes(dtype):
    subs = dtype.__subclasses__()  # <-- Doble guion bajo aquí
    if not subs:
        return dtype
    return [dtype, [subtypes(dt) for dt in subs]]

subtypes(np.generic)
```
