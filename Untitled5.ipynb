# **Seleccion de Filas**

Altura < 100


```python
import pandas as pd
personajesSW = pd.read_csv('/content/characters.csv')
personjesBajosSW = personajesSW[personajesSW["height"]<100]
print('Tamaño del resultado:',personjesBajosSW.shape)
personjesBajosSW
```

altura > 170 o peso > 100kg



```python
import pandas as pd
personajesSW = pd.read_csv('/content/characters.csv')
personjesAltosPesadosSW = personajesSW[(personajesSW["height"]>170) | (personajesSW["mass"]>100)]
print('Tamaño del resultado:',personjesAltosPesadosSW.shape)
personjesAltosPesadosSW
```

Filas por segmentacion


```python
import pandas as pd
personajesSW = pd.read_csv('/content/characters.csv')
personjesAltosPesadosSW = personajesSW[10:40]
print('Tamaño del resultado:',personjesAltosPesadosSW.shape)
personjesAltosPesadosSW
```

# **Filtrado de filas y columnas**

loc altura > 170

```python
import pandas as pd
peronasjesSW = pd.read_csv('/content/characters.csv')
peronajesAltosSW = personajesSW.loc[personajesSW["height"]>170, "name"]
print('Tamaño del resultado:', peronajesAltosSW.shape)
peronajesAltosSW
```

loc altura > 170, nombre y altura



```python
import pandas as pd
personajesSW = pd.read_csv('/content/characters.csv')
personajesAltosSW = personajesSW.loc[personajesSW["height"]>170, ["name","height"]]
print('Tamaño del resultado:', personajesAltosSW.shape)
personajesAltosSW
```
